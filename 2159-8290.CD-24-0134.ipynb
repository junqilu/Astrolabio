# import packages

```python
import requests
import os
import tarfile
import pandas as pd
import json
```

# Download data

```python
url = 'https://www.ncbi.nlm.nih.gov/geo/download/?acc=GSE254379&format=file'
direct_input = 'input/2159-8290.CD-24-0134'

os.makedirs(direct_input, exist_ok=True)
```

```python
with requests.get(url, stream=True) as r:
    r.raise_for_status()
    with open(os.path.join(direct_input, "GSE143423_RAW.tar"), "wb") as f:
        for chunk in r.iter_content(chunk_size=8192):
            if chunk:
                f.write(chunk)
```

```python
direct_tar = os.path.join(direct_input, "GSE143423_RAW.tar")
direct_tar_folder = os.path.join(direct_input, "GSE143423_RAW")

with tarfile.open(direct_tar, "r") as tar:
    tar.extractall(direct_tar_folder, filter='data')

os.remove(direct_tar)
```

```python
input_data_dict = {}

for disease, disease_data_id in {
    'lung_cancer_brain_metastasis': 'GSM8041245',
    'breast_cancer_brain_metastasis': 'GSM8041246'
}.items():
    input_data_dict[disease] = {}

    if disease == 'lung_cancer_brain_metastasis':
        sample_idx = '1'
    else:
        sample_idx = '2'

    for file, direct_file in {
        'barcodes': f'barcodes_S{sample_idx}.tsv.gz',
        'features': f'features_S{sample_idx}.tsv.gz',
        'matrix': f'matrix_S{sample_idx}.mtx.gz'
    }.items():
        input_data_dict[disease][file] = os.path.join(direct_tar_folder, f'{disease_data_id}_{direct_file}')

print(json.dumps(input_data_dict, indent=4))
```
