Tumor-associated reactive astrocytes aid the evolution of immunosuppressive environment in glioblastoma

# Import packages

```python
import requests
import os
import tarfile
import pandas as pd
import json
```

# Download data

```python
url = 'https://www.ncbi.nlm.nih.gov/geo/download/?acc=GSE128536&format=file'
direct_input = 'input/s41467-019-10493-6'

os.makedirs(direct_input, exist_ok=True)
```

```python
with requests.get(url, stream=True) as r:
    r.raise_for_status()
    with open(os.path.join(direct_input, "GSE128536_RAW.tar"), "wb") as f:
        for chunk in r.iter_content(chunk_size=8192):
            if chunk:
                f.write(chunk)
```

```python
direct_tar = os.path.join(direct_input, "GSE128536_RAW.tar")
direct_tar_folder = os.path.join(direct_input, "GSE128536_RAW")

with tarfile.open(direct_tar, "r") as tar:
    tar.extractall(direct_tar_folder, filter='data')

os.remove(direct_tar)
```

```python
from pathlib import Path

raw_dir = Path(r"input\s41467-019-10493-6\GSE128536_RAW")

cortex_tumor_samples = {
    "GSM3679135": "Cortex_1",
    "GSM3679136": "Cortex_2",
    "GSM3679143": "Cortex_3",
    "GSM3679144": "Cortex_4",
    "GSM3679145": "Cortex_5",

    "GSM3679137": "Tumor_6",
    "GSM3679138": "Tumor_1",
    "GSM3679139": "Tumor_2",
    "GSM3679140": "Tumor_3",
    "GSM3679141": "Tumor_4",
    "GSM3679142": "Tumor_5",
    "GSM3679146": "Tumor_7",
}

selected_files = [
    f for f in raw_dir.iterdir()
    if f.is_file() and any(f.name.startswith(gsm) for gsm in cortex_tumor_samples.keys())
]

for f in sorted(selected_files):
    print(f.name)
```

```python
test_file = sorted(selected_files)[0]
df = pd.read_csv(test_file, sep="\t", compression="gzip")
print(df.columns.tolist())
print(df.head(10))
```
